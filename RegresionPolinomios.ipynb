```python
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
cant_datos = 100
def sen(a,b,x):
  return a*np.sin(np.pi*x)+b

data = dict()
data['x'] = np.linspace(-1,1,num=cant_datos)
data['y_real'] = sen(2,1,data['x'])
data['y'] = data['y_real'] + np.random.normal(0,0.5,size=cant_datos)

plt.figure(figsize=(10,7))
plt.scatter(data['x'],data['y'],label='Datos medidos',s=50,alpha=0.5)
plt.plot(data['x'],data['y_real'],color='red',label='Curva real')
plt.legend()
plt.show()
```

```python
# Ajusto con un polinomio de grado = 3
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Se genera un modelo con el objecto polyfeats, el termino de bias es el de x⁰
polyfeats = PolynomialFeatures(degree = 3, include_bias = True)

x = np.array(data['x']).reshape(-1, 1) # sklearn espera matriz
x_poly = polyfeats.fit_transform(x)
# Son las columnas de x⁰, x¹, x², x³
# print(x_poly)

reg = LinearRegression(fit_intercept = False) # False, no ajusta la ordenada a la origen

reg.fit(x_poly, data['y']) # Utilizo la matriz con las potencias de x

print(f"Los coeficientes son {reg.coef_}")

x_aux = np.arange(-1.2,1.2,0.01)
plt.figure()
plt.scatter(data['x'], data['y'])
plt.plot(x_aux, np.polyval(np.flip(reg.coef_), x_aux), color = 'red')
```
